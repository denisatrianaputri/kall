# Sistem Persamaan Linier
## Pengertian
Merupakan sekumpulan persamaan linier yang terdiri dari beberapa variabel,sistem ini terdiri dari tiga persamaan dengan tiga variabel x, y, z. Solusi sistem linear ini adalah nilai yang dapat menyelesaikan persamaan ini.
## Jenis-jenis
Sistem persamaan linier dapat dibedakan menjadi tiga jenis berdasarkan jumlah variabelnya, yaitu:

1. Sistem persamaan linier satu variabel

  Adalah kalimat terbuka yang menggunakan tanda sama dengan (=) dan hanya memiliki satu variabel berpangkat satu. Bentuk umum SPLSV adalah ax + b = 0, dengan a dan b bilangan bulat bukan no
2. Sistem persamaan linier dua variabel

  Adalah dua buah persamaan linier dua variabel yang memiliki satu penyelesaian yang memenuhi kedua persamaan tersebut.
3.  Sistem persamaan linier tiga variabel

  Konsep dalam ilmu matematika yang digunakan untuk menyelesaikan kasus yang tidak dapat diselesaikan menggunakan persamaan linear satu variabel dan persamaan linier dua variabel
