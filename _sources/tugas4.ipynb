# **Eigenvalue dan Eigenvector**
## Definisi Eigenvalue dan Eigenvector

**Eigenvalue** (nilai eigen) adalah skalar $(\lambda)$ yang memenuhi:


$$
[
A\mathbf{v} = \lambda \mathbf{v}
]
$$


dengan:

- $( A )$ adalah matriks kuadrat  $( n \times n )$,
- $( \mathbf{v} \neq \mathbf{0})$ adalah vektor eigen (tak nol),
- $( \lambda)$ adalah nilai eigen.

**Eigenvector** (vektor eigen) adalah vektor tak nol $( \mathbf{v} )$ yang ketika dikalikan dengan matriks $( A )$, menghasilkan:


$$
[
A\mathbf{v} = \lambda \mathbf{v}
]
$$



Artinya, arah $( \mathbf{v} )$ tidak berubah setelah transformasi oleh matriks $( A )$, hanya diperbesar, diperkecil, atau dibalik sesuai nilai $( \lambda )$.

---

**Catatan Penting:**

Untuk mencari nilai eigen, gunakan:



$$
[
\det(A - \lambda I) = 0
]
$$


Solusi $( \lambda )$ dari persamaan ini adalah nilai eigen. Setelah itu, untuk mencari vektor eigen, substitusikan kembali $( \lambda )$ ke dalam:



$$
[
(A - \lambda I)\mathbf{v} = 0
]
$$




## Mencari Eigenvalues dan Eigenvectors Menggunakan Polinomial Karakteristik

# Eigenvalue dan Eigenvektor


## 1. Definisi Eigenvalue dan Eigenvector

Untuk sebuah matriks kuadrat $A$, sebuah eigenvalue $\lambda$ dan eigenvector $\vec{v}$ memenuhi:

$A\vec{v} = \lambda \vec{v}$

dengan syarat:

- $\vec{v} \neq \vec{0}$ (vektor tak nol)
- $\lambda \in \mathbb{R}$ atau $\mathbb{C}$
- $A \in \mathbb{R}^{n \times n}$ atau $\mathbb{C}^{n \times n}$

### Interpretasi Geometris:
Vektor $\vec{v}$ tidak berubah arah ketika dikalikan dengan matriks $A$, hanya panjang dan arahnya yang bisa berubah (karena $\lambda$).



## 2. Mencari Eigenvalue dan Eigenvector dengan Polinomial Karakteristik

Urutkan eigenvector berdasarkan eigenvalue dari besar ke kecil.

### Contoh Soal

### Matriks A:

$$
A :
\begin{bmatrix}
-2 & 0 \\
0 & 4
\end{bmatrix}
$$

### Jawaban:

### Langkah 1: Menentukan Eigenvalue

Gunakan persamaan karakteristik:

$$
\det(A - \lambda I) = 0
$$

$$
A - \lambda I = \begin{bmatrix}
3 - \lambda & 0 \\
0 & 3 - \lambda
\end{bmatrix}
$$

$$
\Rightarrow \det(A - \lambda I) = (3 - \lambda)^2 = 0
$$

$$
\Rightarrow \lambda_1 = 3, \quad \lambda_2 = 3
$$

Nilai eigen $\lambda = 3$ muncul dua kali → aljabar multiplicity = 2

---

### Langkah 2: Menentukan Eigenvector

Kita cari vektor $\vec{v}$ yang memenuhi:

$$
(A - 3I)\vec{v} = \vec{0}
$$

$\Rightarrow \begin{bmatrix}
0 & 0 \\
0 & 0
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
=$ $\begin{bmatrix}
0 \\
0
\end{bmatrix}$


Persamaan ini selalu benar untuk semua vektor tak nol.

$$
\Rightarrow \text{Eigenvector: } \vec{v} =
\begin{bmatrix}
x \\
y
\end{bmatrix}, \quad \text{dengan } x, y \in \mathbb{R},\ (x, y) \neq (0, 0)
$$

---

### Matriks B:

$$
B = \begin{bmatrix}
-2 & 0 \\
0 & 4
\end{bmatrix}
$$

---

### Langkah 1: Mencari Eigenvalue

$$
\det(B - \lambda I) = (-2 - \lambda)(4 - \lambda) = 0
$$

$$
\Rightarrow \lambda_1 = -2,\quad \lambda_2 = 4
$$

---

### Langkah 2: Mencari Eigenvector

#### Untuk $\lambda = -2$:

$$
(B + 2I) =
\begin{bmatrix}
0 & 0 \\
0 & 6
\end{bmatrix}
\Rightarrow y = 0
$$

$$
\vec{v}_1 =
x\begin{bmatrix}
1 \\
0
\end{bmatrix}, \quad x \neq 0
$$

#### Untuk $\lambda = 4$:

$$
(B - 4I) =
\begin{bmatrix}
-6 & 0 \\
0 & 0
\end{bmatrix}
\Rightarrow x = 0
$$

$$
\vec{v}_2 =
y\begin{bmatrix}
0 \\
1
\end{bmatrix}, \quad y \neq 0
$$

---

### Hasil Akhir:

- Eigenvalue: $\lambda = -2, \lambda = 4$
- Eigenvector:
  - Untuk $\lambda = -2$ : $\vec{v}_1 = x$ $\begin{bmatrix} 1 \\ 0 \end{bmatrix}$
  - Untuk $\lambda = 4$ : $\vec{v}_2 = y$ $\begin{bmatrix} 0 \\ 1 \end{bmatrix}$


## 3. urutkan dari besar ke kecil


```python
import numpy as np

# Matriks A
A = np.array([[-2, 0],
              [0, 4]])

# Hitung eigenvalue dan eigenvector
eigenvalues, eigenvectors = np.linalg.eig(A)

# Urutkan indeks berdasarkan eigenvalue dari besar ke kecil
sorted_indices = np.argsort(eigenvalues)[::-1]

# Urutkan eigenvalue dan eigenvector sesuai indeks
eigenvalues_sorted = eigenvalues[sorted_indices]
eigenvectors_sorted = eigenvectors[:, sorted_indices]

# Cetak hasil
print("Eigenvalues (terurut):")
print(eigenvalues_sorted)

print("Eigenvectors (mengikuti Aeigenvalues):")
print(eigenvectors_sorted)
```
