## Pembuktian Transformasi Rotasi di Bidang $$(\mathbb{R}^2)$$

Buktikan bahwa transformasi linier:
$$
T : \mathbb{R}^2 \rightarrow \mathbb{R}^2
$$
dengan matriks:
$$
A = \begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}
$$
merupakan rotasi berlawanan arah jarum jam sebesar sudut \(\theta\).

---

### Langkah-langkah Pembuktian:

Misalkan suatu vektor $$(\vec{v} = (x, y))$$ dalam koordinat polar dapat dituliskan sebagai:
$$
\vec{v} = (r \cos \alpha, r \sin \alpha)
$$
dengan:
- $$(r = \|\vec{v}| = \sqrt{x^2 + y^2})$$: panjang vektor  
- $$(\alpha)$$: sudut antara vektor dan sumbu-X

Lakukan transformasi matriks rotasi terhadap $$(\vec{v}):$$
$$
A \vec{v} =
\begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}
\begin{bmatrix}
r \cos \alpha \\
r \sin \alpha
\end{bmatrix}
$$

Hasil perkalian matriks:
$$
A \vec{v} =
\begin{bmatrix}
r \cos \theta \cos \alpha - r \sin \theta \sin \alpha \\
r \sin \theta \cos \alpha + r \cos \theta \sin \alpha
\end{bmatrix}
= r
\begin{bmatrix}
\cos \theta \cos \alpha - \sin \theta \sin \alpha \\
\sin \theta \cos \alpha + \cos \theta \sin \alpha
\end{bmatrix}
$$

Gunakan identitas trigonometri:
$$
 \cos(\alpha + \theta) = \cos \alpha \cos \theta - \sin \alpha \sin \theta\\
\sin(\alpha + \theta) = \sin \alpha \cos \theta + \cos \alpha \sin \theta)
$$
Sehingga:
$$
A \vec{v} = r
\begin{bmatrix}
\cos(\alpha + \theta) \\
\sin(\alpha + \theta)
\end{bmatrix}
$$

---

### Kesimpulan:

Transformasi oleh matriks \(A\) memutar vektor sebesar sudut $$(\theta)$$ berlawanan arah jarum jam tanpa mengubah panjangnya.  
Maka, terbukti bahwa:
$$
A = \begin{bmatrix}
\cos \theta & -\sin \theta \\
\sin \theta & \cos \theta
\end{bmatrix}
$$
adalah matriks rotasi di bidang $$(\mathbb{R}^2)$$
