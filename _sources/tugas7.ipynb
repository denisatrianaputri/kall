# Cross Produt

## Definsi Hasil Kali Silang

### Misal:
$$
\vec{u} = \begin{bmatrix} u_1 \\ u_2 \\ u_3 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix}
$$

adalah vektor-vektor di $\mathbb{R}^3$. Hasil kali silang dari $\vec{u}$ dan $\vec{v}$, dinotasikan sebagai $\vec{u} \times \vec{v}$, adalah vektor:






$$
\vec{u} \times \vec{v} =
\begin{bmatrix}
u_2 v_3 - u_3 v_2 \\
-(u_1 v_3 - u_3 v_1) \\
u_1 v_2 - u_2 v_1
\end{bmatrix}
$$


### Rumus:

$$
i=(1,0,0)\\
j=(0,1,0)\\
ixj=(0,0,1)=k\\
jxi(0,0,-1)=-k
$$



### Contoh Menghitung Hasil Kali Silang

Misalkan
$\vec{u} = \begin{bmatrix} 2 \\ -1 \\ 4 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 3 \\ 2 \\ 5 \end{bmatrix}$





Temukan $\vec{u} \times \vec{v}$, dan verifikasi bahwa hasil kali silang ini tegak lurus terhadap baik $\vec{u}$ maupun $\vec{v}$.





### Solusi Menggunakan Definisi

$$
\vec{u} \times \vec{v} =
\begin{bmatrix}
(-1)(5) - (4)(2) \\
-((2)(5) - (4)(3)) \\
(2)(2) - (-1)(3)
\end{bmatrix}
=
\begin{bmatrix}
-5 - 8 \\
-(10 - 12) \\
4 + 3
\end{bmatrix}
=
\begin{bmatrix}
-13 \\
2 \\
7
\end{bmatrix}
$$



### Verifikasi Vektor Tegak Lurus

Diketahui hasil kali silang:




$$
\vec{u} \times \vec{v} = \begin{bmatrix} -13 \\ 2 \\ 7 \end{bmatrix}
$$



dan



$$
\vec{u} = \begin{bmatrix} 2 \\ -1 \\ 4 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 3 \\ 2 \\ 5 \end{bmatrix}
$$

### Cek dot product:



$$
(\vec{u} \times \vec{v}) \cdot \vec{u} =
(-13)(2) + (2)(-1) + (7)(4) = -26 - 2 + 28 = 0
$$





$$
(\vec{u} \times \vec{v}) \cdot \vec{v} =
(-13)(3) + (2)(2) + (7)(5) = -39 + 4 + 35 = 0
$$



Karena hasil dot product keduanya adalah 0, maka:  
*$\vec{u} \times \vec{v}$* tegak lurus terhadap *$\vec{u}$* dan *$\vec{v}$*


### Menghitung Hasil Kali Silang Menggunakan Determinan

Pertama, kita membentuk array $3 \times 3$ seperti yang ditunjukkan di bawah ini:





$$
\begin{vmatrix}
\vec{i} & \vec{j} & \vec{k} \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3 \\
\end{vmatrix}
$$






$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
u_2 & u_3 \\
v_2 & v_3 \\
\end{vmatrix} \vec{i} -
\begin{vmatrix}
u_1 & u_3 \\
v_1 & v_3 \\
\end{vmatrix} \vec{j} +
\begin{vmatrix}
u_1 & u_2 \\
v_1 & v_2 \\
\end{vmatrix} \vec{k}
$$




$$
= (u_2v_3 - u_3v_2)\vec{i} - (u_1v_3 - u_3v_1)\vec{j} + (u_1v_2 - u_2v_1)\vec{k}
$$






$$
\vec{u} = \begin{bmatrix} 2 \\ -1 \\ 4 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 3 \\ 2 \\ 5 \end{bmatrix}
$$




Gunakan determinan dari matriks:



$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & -1 & 4 \\
3 & 2 & 5
\end{vmatrix}
$$




Hitung masing-masing komponen:

$$
\hat{i} = (-1)(5) - (4)(2) = -5 - 8 = -13
$$




$$
\hat{j} = - \left[ (2)(5) - (4)(3) \right] = - (10 - 12) = 2
$$



$$
\hat{k} = (2)(2) - (-1)(3) = 4 + 3 = 7
$$




Jadi:




$$
\vec{u} \times \vec{v} = -13\hat{i} + 2\hat{j} + 7\hat{k} =
\begin{bmatrix}
-13 \\
2 \\
7
\end{bmatrix}
$$




### Sifat-Sifat Perkalian Silang


1. $\vec{u} \times \vec{v} = -(\vec{v} \times \vec{u})$ *Anticommutative Property*

2. a. $(\vec{u} + \vec{v}) \times \vec{w} = \vec{u} \times \vec{w} + \vec{v} \times \vec{w}$ *Distributive Properties*  
   b. $\vec{u} \times (\vec{v} + \vec{w}) = \vec{u} \times \vec{v} + \vec{u} \times \vec{w}$

3. $c(\vec{u} \times \vec{v}) = (c\vec{u}) \times \vec{v} = \vec{u} \times (c\vec{v})$

4. a. $(\vec{u} \times \vec{v}) \cdot \vec{u} = 0$ *Orthogonality Properties*  
   b. $(\vec{u} \times \vec{v}) \cdot \vec{v} = 0$

5. $\vec{u} \times \vec{u} = \vec{0}$

6. $\vec{u} \times \vec{0} = \vec{0}$

7. $\vec{u} \cdot (\vec{v} \times \vec{w}) = (\vec{u} \times \vec{v}) \cdot \vec{w}$ *Triple Scalar Product*





# Hasil Kali Silang Dan Sudut


Misalkan $\vec{u}$ dan $\vec{v}$ adalah vektor tak nol di $\mathbb{R}^3$. Maka




$$
\|\vec{u} \times \vec{v}\| = \|\vec{u}\| \|\vec{v}\| \sin(\theta),
$$




di mana $\theta$, $0 \leq \theta \leq \pi$, adalah sudut antara $\vec{u}$ dan $\vec{v}$.


## Latian



$$
\vec{u} = \begin{bmatrix} 1 \\ 3 \\ 6 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} -1 \\ 2 \\ 1 \end{bmatrix}
$$




Hitung $\|\vec{u} \times \vec{v}\|$

Langkah-langkah:



$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\vec{i} & \vec{j} & \vec{k} \\
1 & 3 & 6 \\
-1 & 2 & 1
\end{vmatrix}
= (3 \cdot 1 - 6 \cdot 2)\vec{i} - (1 \cdot 1 - 6 \cdot (-1))\vec{j} + (1 \cdot 2 - 3 \cdot (-1))\vec{k}
$$




$$
= -9\vec{i} - 7\vec{j} + 5\vec{k}
= \begin{bmatrix} -9 \\ -7 \\ 5 \end{bmatrix}
$$



$$
\|\vec{u} \times \vec{v}\| = \sqrt{(-9)^2 + (-7)^2 + 5^2} = \sqrt{81 + 49 + 25} = \sqrt{155}
$$




# Implementasi



Secara geometri luas jajaran genjang adalah $A = bh$, di mana $b$ adalah panjang alas dan $h$ adalah tinggi jajaran genjang, seperti yang ditunjukkan dalam **Gambar (a)**.

Ketika kita mendefinisikan **aturan jajaran genjang penjumlahan vektor**, dua vektor $\vec{u}$ dan $\vec{v}$ membentuk jajaran genjang jika ditarik dari titik awal yang sama, seperti yang ditunjukkan dalam **Gambar (b)**.

Trigonometri mengatakan bahwa $h = \|\vec{u}\| \sin(\theta)$, sehingga luas jajaran genjang adalah:




$$
A = \|\vec{u}\| \|\vec{v}\| \sin(\theta) = \|\vec{u} \times \vec{v}\|
$$

### Volume Paralelepiped

Bangun ruang tiga dimensi yang dibentuk oleh enam buah jajar genjang.

Dengan menghitung hasil kali silang $\vec{v}$ dan $\vec{w}$, kita mendapatkan vektor yang besarnya sama dengan luas alas. Menghitung produk titik vektor ini dengan $\vec{u}$ menghitung volume paralelepiped!

Oleh karena itu, **volume paralelepiped yang ditentukan oleh vektor** $\vec{u}$, $\vec{v}$, dan $\vec{w}$ **adalah:**

$$
V = |\vec{u} \cdot (\vec{v} \times \vec{w})|
$$

### Contoh: Menghitung Volume Paralelepiped

Tentukan volume paralelepiped yang ditentukan oleh vektor:

$$
\vec{u} = \begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} -1 \\ 1 \\ 0 \end{bmatrix}, \quad
\text{dan} \quad
\vec{w} = \begin{bmatrix} 0 \\ 1 \\ 1 \end{bmatrix}
$$

Pertama, kita hitung:

$$
\vec{v} \times \vec{w} =
\begin{bmatrix} 1 \\ 1 \\ -1 \end{bmatrix}
$$

Kemudian,

$$
|\vec{u} \cdot (\vec{v} \times \vec{w})| =
\begin{bmatrix} 1 \\ 1 \\ 0 \end{bmatrix} \cdot
\begin{bmatrix} 1 \\ 1 \\ -1 \end{bmatrix} = 2
$$

Oleh karena itu, volume paralelepiped adalah 2 satuan kubik.

## Penyelesaian Soal
#  Jawaban Soal Perkalian Silang (Cross Product)

Jika





$$
\vec{u} = \begin{bmatrix} u_1 \\ u_2 \\ u_3 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix}
$$




Maka:



$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
u_1 & u_2 & u_3 \\
v_1 & v_2 & v_3 \\
\end{vmatrix}
= (u_2v_3 - u_3v_2)\hat{i} - (u_1v_3 - u_3v_1)\hat{j} + (u_1v_2 - u_2v_1)\hat{k}
$$






---

##  Soal 1




$$
\vec{u} = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 2 \\ 1 \end{bmatrix}
$$

Ubah ke dimensi 3:




$$
\vec{u} = \begin{bmatrix} 1 \\ 2 \\ 0 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 2 \\ 1 \\ 0 \end{bmatrix}
$$






$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
1 & 2 & 0 \\
2 & 1 & 0 \\
\end{vmatrix}
= 0\hat{i} - 0\hat{j} -3\hat{k}
$$







$$
|\vec{u} \times \vec{v}| = 3
$$

---

##  Soal 2



$$
\vec{u} = \begin{bmatrix} 2 \\ 0 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 0 \\ 3 \end{bmatrix}
$$

Ubah ke dimensi 3:



$$
\vec{u} = \begin{bmatrix} 2 \\ 0 \\ 0 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 0 \\ 3 \\ 0 \end{bmatrix}
$$





$$
\vec{u} \times \vec{v} =
\begin{vmatrix}
\hat{i} & \hat{j} & \hat{k} \\
2 & 0 & 0 \\
0 & 3 & 0 \\
\end{vmatrix}
= 0\hat{i} - 0\hat{j} + 6\hat{k}
$$






$$
|\vec{u} \times \vec{v}| = 6
$$

##  Soal 3
