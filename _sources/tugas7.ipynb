#**Cross Produt**

##Definsi Hasil Kali Silang

###Misal:
$$
\vec{u} = \begin{bmatrix} u_1 \\ u_2 \\ u_3 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} v_1 \\ v_2 \\ v_3 \end{bmatrix}
$$

adalah vektor-vektor di ${R}^3$. Hasil kali silang dari $\vec{u}$ dan $\vec{v}$, dinotasikan sebagai $\vec{u} \times \vec{v}$, adalah vektor:






$$
\vec{u} \times \vec{v} =
\begin{bmatrix}
u_2 v_3 - u_3 v_2 \\
-(u_1 v_3 - u_3 v_1) \\
u_1 v_2 - u_2 v_1
\end{bmatrix}
$$




### Contoh Menghitung Hasil Kali Silang

Misalkan
$\vec{u} = \begin{bmatrix} 2 \\ -1 \\ 4 \end{bmatrix}, \quad
\vec{v} = \begin{bmatrix} 3 \\ 2 \\ 5 \end{bmatrix}$





Temukan $\vec{u} \times \vec{v}$, dan verifikasi bahwa hasil kali silang ini tegak lurus terhadap baik $\vec{u}$ maupun $\vec{v}$.





#### Solusi Menggunakan Definisi

$$
\vec{u} \times \vec{v} =
\begin{bmatrix}
(-1)(5) - (4)(2) \\
-((2)(5) - (4)(3)) \\
(2)(2) - (-1)(3)
\end{bmatrix}
=
\begin{bmatrix}
-5 - 8 \\
-(10 - 12) \\
4 + 3
\end{bmatrix}
=
\begin{bmatrix}
-13 \\
2 \\
7
\end{bmatrix}
$$





```python
import numpy as np

v = np.array([3, 2, 5])
u = np.array([2, -1, 4])

cross = np.cross(v, u)
print("Hasil cross product u × v:", cross)
```
